```python
import sys 
sys.path.append('..')
import matplotlib.pyplot as plt
from glob import glob
import ra
import numpy as np
import pylab 
import scipy.stats as stats
from utils import *

def show_images(images, cols = 1):
    n_images = len(images)
    fig = plt.figure()
    for n, image in enumerate(images):
        a = fig.add_subplot(cols, np.ceil(n_images/float(cols)), n + 1)
        plt.imshow(image, cmap='gray')
    fig.set_size_inches(np.array(fig.get_size_inches()) * n_images)
    plt.show()
```

```python
a = np.load('./data/recon_fft_new/face_orig.npy')
b = np.load('./data/recon_fft_new/face_pixel.npy')
c = np.load('./data/recon_fft_new/face_fft_2.npy')
show_images([a, b, c])
```
