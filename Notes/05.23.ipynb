```python
import sys 
sys.path.append('..')
import matplotlib.pyplot as plt
from glob import glob
import ra
import numpy as np
import pylab 
import scipy.stats as stats
from utils import *

def show_images(images, cols = 1):
    n_images = len(images)
    fig = plt.figure()
    for n, image in enumerate(images):
        a = fig.add_subplot(cols, np.ceil(n_images/float(cols)), n + 1)
        plt.imshow(image, cmap='gray')
    fig.set_size_inches(np.array(fig.get_size_inches()) * n_images)
    plt.show()
```

```python
# fft histograms of three different test mri images

data = glob("../data/valid_img_slices/*.ra")
np.random.shuffle(data)
data = data[0:3]
sample = [get_image_old2(s) for s in data]
show_images(sample)
y = np.reshape(sample[0], -1)
y2 = np.reshape(sample[1], -1)
y3 = np.reshape(sample[2], -1)
m_ = np.shape(y)[0]
fig = plt.figure()
a = fig.add_subplot(1, 3, 1)
plt.plot(y)
plt.plot(y2)
b = fig.add_subplot(1, 3, 2)
plt.plot(y2)
plt.plot(y3)
c = fig.add_subplot(1, 3, 3)
plt.plot(y3)
plt.plot(y)
fig.set_size_inches(np.array(fig.get_size_inches()) * 3)
plt.show()
```
